## Example - Health Survey

```python
survey_raw = pl.read_csv("./data/health_survey.csv")
survey_raw.columns
```

## Renaming columns

* Use `rename` method to rename one or more columns.
* Use a `dict` of pair, i.e. `{'old_name_1':'new_name_1', ...}`

```python
survey = survey_raw.rename({'':'ID'})
survey.head()
```

## Renaming many columns with a `dict` comprehension

```python
fix_name = lambda n: n.lower().replace('.', '_')

{n:fix_name(n) for n in survey.columns}
```

```python
survey = (survey_raw 
          .rename({'':'ID'})
          .rename({n:fix_name(n) for n in survey_raw.columns}) 
         )
survey.head(2)
```

## Dropping one or more columns with `drop`

#### Drop a single column

```python
(survey
 .drop('f1')
 .head(2))
```

#### Dropping a list of multiple columns

```python
(survey
 .drop(['f1', 'f2'])
 .head(2))
```

## `select` by column `dtype`

```python

(survey
 .select(pl.col(pl.Utf8))
 .head(2)
)
```

## `select` using regular expression

We can use `pl.col('reg_ex')`, provided the `reg_ex` starts with `^` and ends with `$` This covers the following use cases.

* `starts_with`
* `ends_with`
* `contains`

#### Columns starting with `f1`

```python
(survey
 .select(pl.col('^f1.*$'))
 .head(2)
)
```

#### Columns starting with `f1` of `f2`

```python
(survey
 .select(pl.col('^(f1|f2).*$'))
 .head(2)
)
```

#### Columns containing `_`

```python
(survey
 .select(pl.col('^.*_.*$'))
 .head(2)
)
```

## `exclude` column

Unlike `dfply` and `pyspark`, the `~` doesn't negate, but we can use the `exclude` method instead

We can use `pl.all().exclude('reg_ex')`, provided the `reg_ex` starts with `^` and ends with `$` 

#### No `f1`

```python
(survey
 .select(pl.all().exclude('^f1.*$'))
 .head(2))
```

#### No `_`

```python
(survey
 .select(pl.all().exclude('^.*_.*$'))
 .head(2))
```

#### Excluding columns by type

```python
(survey
 .select(pl.all().exclude(pl.Int64))
 .head(2))
```

## Pipeable `select` helper based position

* `cols.startswith(substr)`
* `cols.endswith(substr)`
* `cols.contains(substr)`

```python
from more_polars import cols

(survey
 .select(cols.startswith('f1'))
 .head(2)
)
```

```python
(survey
 .select(cols.endswith('_2'))
 .head(2)
)
```

```python
(survey
 .select(cols.contains('_'))
 .head(2)
)
```

## `select` helper based position

* `df.columns >> cols.between(start_col, end_col, inclusive=True)`
* `df.columns >> cols.to(end_col)`
* `df.columns >> cols.from_(start_col)`

```python
(survey
 .select(survey.columns >> cols.between('f1', 'f1_1'))
 .head(2))
```

```python
(survey
 .select(survey.columns >> cols.between('f1', 'f1_1', inclusive=False))
 .head(2))
```

```python
(survey
 .select(survey.columns >> cols.from_('f5_7'))
 .head(2))
```

```python
(survey
 .select(survey.columns >> cols.to('f1_1'))
 .head(2))
```

#### Composing helpers in a `list`

```python
(survey
 .select([cols.startswith('f1'),
          cols.startswith('f2'),
         ])
 .head(2))
```

#### Beware of duplicate columns!

Unfortunately, duplicate columns will raise an exception.

```python
(survey
 .select([cols.startswith('f1'),
          cols.contains('_'),
         ])
 .head(2))
```
